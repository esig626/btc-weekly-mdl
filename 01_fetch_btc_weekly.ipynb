```python
import pandas as pd
import yfinance as yf

#cd /Users/eliassiguenza/Desktop/Sim/Finance
# source .venv/bin/activate

#
# Download BTC-USD weekly OHLCV data from Yahoo Finance
df = yf.download(
    tickers="BTC-USD",
    interval="1wk",
    auto_adjust=False,
    progress=False
)

# Basic clean-up
df = df.dropna()
df.index.name = "Date"

# Display a quick summary
summary = pd.DataFrame({
    "rows": [df.shape[0]],
    "cols": [df.shape[1]],
    "start": [df.index.min()],
    "end": [df.index.max()],
    "columns": [", ".join(df.columns)]
})

print(summary.to_string(index=False))
print("\nHead:")
display(df.head(5))
print("\nTail:")
display(df.tail(5))

```
